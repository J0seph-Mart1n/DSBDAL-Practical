Printing columns with null values

```python
cols_with_na = []
for col in df.columns:
    if df[col].isnull().any():
        cols_with_na.append(col)

print(cols_with_na)
```

Replacing Null values with mean value of column

```python
df["Attendance"] = df["Attendance"].fillna(df["Attendance"].mean())
df["Sub1"] = df["Sub1"].fillna(df["Sub1"].mean())
df["Sub2"] = df["Sub2"].fillna(df["Sub2"].mean())
df["Sub3"] = df["Sub3"].fillna(df["Sub3"].mean())
df["Sub4"] = df["Sub4"].fillna(df["Sub4"].mean())
```

```python
df.isnull().sum()
```

```python
df.dropna(axis=1)
```

Detecting outliers using Inter Quantile Range (IQR):

```python
Q1 = df["Percentage"].quantile(0.25)
Q3 = df["Percentage"].quantile(0.75)
IQR = Q3 - Q1

Lower_limit = Q1 - 1.5 * IQR
Upper_limit = Q3 + 1.5 * IQR

print(f'Q1 = {Q1}, Q3 = {Q3}, IQR = {IQR}, Lower_limit = {Lower_limit}, Upper_limit = {Upper_limit}')
```

```python
df[(df['Percentage'] < Lower_limit) | (df['Percentage'] > Upper_limit)]
```
