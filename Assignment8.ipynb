Distribution Plot

```python
sns.distplot(ds['fare'])
```

```python
sns.displot(ds['fare'], kde=False)
```

Histogram

```python
sns.distplot(ds['fare'], kde=False, bins=10)
```

Joint Plot

```python
sns.jointplot(x='age', y='fare', data=ds)
```

```python
sns.jointplot(x='age', y='fare', data=ds, kind='hex')
```
